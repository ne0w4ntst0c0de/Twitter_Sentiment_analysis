## Stemming

- Stemming is the process of producing morphological variants of a root/base word. 
- eg: Stemming programs are commonly referred to as stemming algorithms or stemmers. A stemming algorithm reduces the words “chocolates”, “chocolatey”, and “choco” to the root word, “chocolate” and “retrieval”, “retrieved”, “retrieves” reduce to the stem “retrieve”.

```python
port_steam = PorterStemmer()
```

- Removing the characters which are not in the range of a-z, A-Z
- Changinng the format of tweets to lowercase
- Looking for the stopwords in the tweets and removing them

```python
def stemming(content):
    steamed_content = re.sub('[^a-zA-z]', ' ', content)
    steamed_content = steamed_content.lower()
    steamed_content = steamed_content.split()
    steamed_content = [port_steam.stem(word) for word in steamed_content if not word in stopwords.words('english')]
    steamed_content = ' '.join(steamed_content)
    
    return steamed_content
```

```python
twitter_data['steamed_content'] = twitter_data['text'].apply(stemming)
```

```python
twitter_data.head()
```

```python
# Seperating the data(Tweets) and lable(Target)

X = twitter_data['steamed_content'].values
Y = twitter_data['Target'].values
```

```python
X
```

```python
Y
```

## Splitting the testing data and training data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# Converting the textual data into numerical data using vectorizer

vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
print(X_train[:10])
```

## Training the Machine Learning Model

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

## Model Evaluation

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy :', training_data_accuracy*100)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy :', testing_data_accuracy*100)
```
